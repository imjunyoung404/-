```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

df = pd.read_csv("/content/diabetes.csv", index_col=0)
df
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = df.drop(columns=['Outcome', 'BMI'])
y = df['BMI']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Dense Layer 기반 신경망 회귀 모델
# 신경망 모델 생성 (Dense Layer만 사용)
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))  # 첫 번째 레이어, 입력 크기 맞춤
model.add(Dense(32, activation='relu'))  # 두 번째 Dense Layer
model.add(Dense(16, activation='relu'))  # 세 번째 Dense Layer
model.add(Dense(1))  # 출력 레이어 (회귀 문제이므로 출력층은 하나)
```

```python
# 모델 컴파일
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# 모델 학습
history = model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# 테스트 세트에 대한 평가
y_pred = model.predict(X_test_scaled)
```

```python
# Dense Layer 신경망 MSE 출력
print(f"Dense Neural Network MSE: {mean_squared_error(y_test, y_pred):.4f}")
```
